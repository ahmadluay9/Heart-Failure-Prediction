# Model Loading

```python
import pandas as  pd
import json
import pickle
```

```python
# Loading Model Files

with open('best_model.pkl', 'rb') as file_1:
  best_model = pickle.load(file_1)

with open('Drop_Columns.txt','r') as file_2:
  Drop_Columns = json.load(file_2)
```

```python
# data inference
df_inf = pd.read_csv('inference.csv')
df_inf_copy = df_inf.copy()
df_inf_copy = df_inf.drop(['Unnamed: 0'],axis=1)
df_inf_copy = df_inf_copy.drop(Drop_Columns,axis=1).sort_index()
# Replace values in the DEATH_EVENT column
df_inf_copy["DEATH_EVENT"] = df_inf_copy["DEATH_EVENT"].replace({0: "No", 1: "Yes"})
df_inf_copy
```

```python
# Predict using Best Model

y_pred_inf = best_model.predict(df_inf_copy) #best_model from load files
y_pred_inf
```

```python
# Create DEATH_EVENT Prediction DataFrame
y_pred_inf_df = pd.DataFrame(y_pred_inf, columns=['DEATH_EVENT_Prediction'],index=df_inf_copy.index)
# Replace values in the DEATH_EVENT column
y_pred_inf_df["DEATH_EVENT_Prediction"] = y_pred_inf_df["DEATH_EVENT_Prediction"].replace({0: "No", 1: "Yes"})
y_pred_inf_df
```

```python
# Comparing the real default_payment_next_month and predicted classification
model_inf = pd.concat([df_inf_copy['DEATH_EVENT'], y_pred_inf_df['DEATH_EVENT_Prediction']], axis=1).sort_index()
model_inf
```

It can be seen that our model has 90% accuracy when tested on data inference.  On the other hand, it should be noted that this model inference is just 10 data, which could be not enough to represent the actual accuracy.
